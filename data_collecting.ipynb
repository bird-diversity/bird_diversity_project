```python
import pandas as pd
```

```python
path = '0001215-260221153910048.csv' #Original Dataset from eBird site
df = pd.read_csv(path, sep='\t')
```

```python
pd.set_option('display.max_columns', None)
df.head()
```

```python
df.isna().sum()
```

```python
#Drop consttant columns
x = df.nunique()
df_cleaned = df.drop(columns=[i for i in df.columns if x[i] <= 1])
```

```python
#Drop redundent columns
df_cleaned = df_cleaned.drop(columns=['day', 'month', 'year'])
```

```python
#Drop unrelated columns
df_cleaned = df_cleaned.drop(columns=['gbifID', 'occurrenceID', 'locality',
                                       'catalogNumber', 'lastInterpreted', 'collectionCode', 'issue',
                                       'order', 'family', 'genus', 'scientificName', 'taxonRank'
                                        ,'taxonKey', 'speciesKey', 'recordedBy'])
```

```python
df_cleaned.head()
```

```python
#Merge with dataset containing stalite image based ratience and cloud coverage dataset
df1 = pd.read_csv('TRUSTED_VIIRS_ALAN_Data.csv')

df2 = df_cleaned.copy()

if 'index' not in df2.columns:
    df2 = df2.reset_index()

print("2. Merging datasets together...")
final_df = pd.merge(df2, df1, on='index', how='left')

print("Success! Your master dataset is completely built and ready for ecological analysis.")
```
